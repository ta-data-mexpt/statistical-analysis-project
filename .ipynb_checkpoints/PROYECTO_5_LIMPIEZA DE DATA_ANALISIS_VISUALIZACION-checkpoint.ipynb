# Regresi√≥n lineal

Ahora se har√° una regresi√≥n lineal donde tambien por medio de la R-square y del Pvalue podremos saber si la variable
de OverallQual afecta la variable del precio

```python
X = sm.add_constant(data_numerico['OverallQual']) 
Y = data_numerico['SalePrice']

model = sm.OLS(Y, X).fit()
predictions = model.predict(X) 

model.summary()
```

```python
#En este ejercicio mi R al cuadrado esta arriba del .5 por lo tanto la curva si se ajusta a los datos 
#Como el P-Value es muy peque√±o rechazamos la hip√≥tesis nula : Prob (F-statistic):2.19e-313
#Lo que podemos deducir es que la variable, en este caso el OverallQual, si afecta el precio
```

## Modelo de predicci√≥n 

* Anteriormente por medio de la correlaci√≥n podemos definir las siguientes variables como las m√°s relevantes que afectan el costo, por lo que ser√°n las variables que usaremos para los modelos 

```python
#r-ajustada penaliza las columnas que no aportan
#ahora haremos una regresesi√≥n lineal para determinar otra forma de sacar
#los intervalos de costo

x = data_prueba[['ExterQual','BsmtQual','FullBath','KitchenQual','GrLivArea','TotalBsmtSF','TotRmsAbvGrd','1stFlrSF','GarageCars',]]
y = data_prueba.SalePrice
```

```python
#Crear nuestros Datasets de entrenamiento y de prueba
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size =.2, random_state=77)
```

```python
print(len(x))
print(len(y))
print(len(x_train))
print(len(x_test))
print(len(y_train))
print(len(y_test))
```

```python
#Modelo de regresi√≥n lineal

linreg = LinearRegression()
linreg.fit(x_train, y_train)
```

```python
coeff_df = pd.DataFrame(linreg.coef_, x.columns, columns=['Coefficient'])
coeff_df
```

```python
#Obtener nuestro coeficiente de Determination R^2
train_score = linreg.score(x_train,y_train)
test_score = linreg.score(x_test,y_test)
```

```python
train_score
```

```python
predicted = linreg.predict(x_test)
```

```python
#Con r2_score podemos calcular el error de nuestro modelo

r2_score(y_test,predicted)
```

```python
data_categorias_pred = pd.DataFrame({'Actual': y_test, 'Predicted':predicted})
```

```python
data_categorias_pred
```

## Ahora lo hacemos con modelo Ridge 

El objetivo de hacerlo con otro modelo es para podeer comprobar el error 
de nuestra predicci√≥n a pesar de haberlo intentado con distintos modelos 

```python
#Regresi√≥on Ridge
rr = Ridge()
rr.fit(x_train,y_train)
```

```python
rr100 = Ridge(alpha=100)
rr100.fit(x_train,y_train)
```

```python
#Obtenemos R^2 de nuestros Ridge
ridge_train_score = rr.score(x_train,y_train)
ridge_test_score = rr.score(x_test,y_test)
ridge_train_score100 = rr100.score(x_train,y_train)
ridge_test_score100 = rr100.score(x_test,y_test)
```

```python
predicted_ridge = rr100.predict(x_test)
```

```python
#Con r2_score podemos calcular el error de nuestro modelo

r2_score(y_test,predicted_ridge)
```

```python
datacateg_predic_ridge = pd.DataFrame({'Actual':y_test, 'Predicted':predicted_ridge})
datacateg_predic_ridge
```

```python
print(train_score)
print(test_score)
print(ridge_train_score)
print(ridge_test_score)
print(ridge_train_score100)
print(ridge_test_score100)
```

```python
#Train Score alto significa que mis datos se ajustan a las datos de la data
#Test score bajo significa que hay Overfitting
```

```python
fig = plt.figure(facecolor='w',figsize=(12, 12))
#Agui graficamos los coeficientes de Ridge con Alpha de 1
plt.plot(rr.coef_,alpha=0.7,linestyle='none',marker='*',markersize=9,\
         color='red',label=r'Ridge; $\alpha = 0.01$',zorder=7)
#Aqui graficamos los coeficientes de Ridge con Alpha de 100
plt.plot(rr100.coef_,alpha=0.5,linestyle='none',marker='d',markersize=10,\
         color='blue',label=r'Ridge; $\alpha = 100$')
#Aqui graficamos los coeficientes de Regresi√≥n lineal
plt.plot(linreg.coef_,alpha=0.4,linestyle='none',marker='o',markersize=14,\
         color='green',label='Linear Regression')
plt.xlabel('Coefficient Index',fontsize=16)
plt.ylabel('Coefficient Magnitude',fontsize=16)
plt.legend(fontsize=13,loc=4)
plt.show()
```

```python
#aqui podemos observar que los datos de mi regresi√≥n lineal, estamuy cerca
#de los datos con alpha 0.01
```
