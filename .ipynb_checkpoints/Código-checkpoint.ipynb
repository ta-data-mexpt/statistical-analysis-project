# 3. Hypothesis testing

H0: El IDS no influye en la presencia del delito de género en las alcaldías de la CDMX.
<br>H1: El IDS influye en la presencia del delito de género en las alcaldías de la CDMX.

```python
print(delitos_ind.var()['Porcentaje'])
print(delitos_ind.var()['IDS'])
```

Dado que las varianzas no son iguales, se debe proceder a aplicar una prueba que tome esto en cuenta: un Welch's Test

```python
ttest_ind(delitos_ind['Porcentaje'], delitos_ind['IDS'], equal_var = False)
```

Aunque un alto índice de desarrollo podría asociarse con menor presencia de delito, el pvalue de nuestra prueba de hipótesis es demasiado alto, lo cual indica que no podemos rechazar la H0.

# 5. Regression
Regresión lineal delitos - indicadores de desarrollo

```python
X = sm.add_constant(delitos_ind.Rezago_educativo)
Y = delitos_ind.Porcentaje

model = sm.OLS(Y,X).fit()
predictions= model.predict(X)

print_model = model.summary()
print(print_model)
```

```python
X = sm.add_constant(delitos_ind.Calidad_espacio_vivienda)
Y = delitos_ind.Porcentaje

model = sm.OLS(Y,X).fit()
predictions= model.predict(X)

print_model = model.summary()
print(print_model)
```

```python
X = sm.add_constant(delitos_ind.Bienes_durables)
Y = delitos_ind.Porcentaje

model = sm.OLS(Y,X).fit()
predictions= model.predict(X)

print_model = model.summary()
print(print_model)
```

```python
X = sm.add_constant(delitos_ind.Adecuacion_Sanitaria)
Y = delitos_ind.Porcentaje

model = sm.OLS(Y,X).fit()
predictions= model.predict(X)

print_model = model.summary()
print(print_model)
```

```python
X = sm.add_constant(delitos_ind.IDS)
Y = delitos_ind.Porcentaje

model = sm.OLS(Y,X).fit()
predictions= model.predict(X)

print_model = model.summary()
print(print_model)
```

# Conclusión general
Ninguno de los modelos de regresión muestra un buen ajuste. Los indicadores que conforman el Índice de Desarrollo de la Ciudad de México y el Índice en sí mismo no tienen ninguna representatividad o significancia respecto a la presencia de delitos en la alcaldías de la ciudad. Por tanto, no se puede proceder a rechazar la hipótesis nula y no podemos tomar estos datos para hacer predicciones asertivas de la presencia del delito en las alcaldías de la ciudad.
