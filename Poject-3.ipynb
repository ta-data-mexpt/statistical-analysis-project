```python
import random

for col in nulls_2.index:
    #Si hay NaN's en strings, busco la moda para reemplazar y si hay más de una moda, elijo una random entre ellas
    if train_2[col].dtype =='object':
        replace = str(train_2[col].mode()[random.randint(0, (len(train_2[col].mode())-1))])
    else:
    #Si hay NaN's en enteros o floats, lo reemplazo con su media
        replace = float(train_2[col].mean())
    train_2[col] = train_2[col].fillna(replace)
```

### Prueba de hipótesis t para variables numéricas

```python
#Función que toma 1000 muestras con un tamaño de 100 y regresa las medias de cada muestra
def generate_mean(data):
    means = list()
    for i in range(1000):
        samples = np.random.choice(data,100, replace = True)
        means.append(samples.mean())
    return means
```

```python
# Pruebas de hipotesis

houses = train_2['SalePrice']
new_houses = train_2['SalePrice'][train_2['YearBuilt'] > 2000]
old_houses = train_2['SalePrice'][train_2['YearBuilt'] <= 2000]

means_houses = generate_mean(houses)
means_new = generate_mean(new_houses)
means_old = generate_mean(old_houses)

print(f'Precio promedio de todas las casas: ${round(houses.mean(),2)} USD')
print(f'Precio promedio de casas construidas después del año 2000: ${round(new_houses.mean(),2)} USD')
print(f'Precio promedio de casas construidas antes del año 2000: ${round(old_houses.mean(),2)} USD')
```

### Prueba de hipótesis para variables categóricas

```python
values = ['Neighborhood','RoofMatl','RoofStyle']

for i in range(len(values)):
    if i != len(values)-1:
        model = ols(f'SalePrice ~ C({values[i]})', data = train_2).fit()
        anova_table = sm.stats.anova_lm(model, typ=2)
        print(f"{anova_table}\n")
    else:
        model = ols(f'SalePrice ~ C({values[i]})', data = train_2).fit()
        anova_table = sm.stats.anova_lm(model, typ=2)
        print(f"{anova_table}\n")
        
        all_values = str()
        
        for i in values:
            if i != values[-1]:
                all_values += (f"C({i})+")
            else:
                all_values += (f"C({i}")
                
        model = ols(f'SalePrice ~ {all_values})', data = train_2).fit()
        anova_table = sm.stats.anova_lm(model, typ=2)
        print(f"{anova_table}\n")
```
