```python
print('Is the data stationary?')

dftest= adfuller(ftd_gld['US dollars per troy'],autolag= 'AIC')

print(f'P-value={dftest[1]}')
print(f'Critical Values:')
for k, v in dftest[4].items():
    print('\t{}:{} The data are {} stationary with {}% of confidence'.format(k, v, "not" if v < dftest[0] else "", 100-int(k[:-1])))
```

```python
adfuller(ftd_gld['US dollars per troy'],autolag=None)[1]
```

```python
def kpss_test(timeseries):
    print ('Results of KPSS Test:')
    kpsstest = kpss(timeseries, regression='c')
    kpss_output = pd.Series(kpsstest[0:3], index=['Test Statistic','p-value','Lags Used'])
    for key,value in kpsstest[3].items():
     kpss_output['Critical Value (%s)'%key] = value
    print (kpss_output)
```

```python
kpss_test(ftd_gld['US dollars per troy'])
```

```python
def test_stationarity(timeseries):
    
    #Determing rolling statistics
    
    rolmean = pd.Series(timeseries).rolling(window=12).mean()
    rolstd = pd.Series(timeseries).rolling(window=12).std()

    #Plot rolling statistics:
    orig = plt.plot(timeseries, color='blue',label='Original')
    mean = plt.plot(rolmean, color='red', label='Rolling Mean')
    std = plt.plot(rolstd, color='black', label = 'Rolling Std')
    plt.legend(loc='best')
    plt.title('Rolling Mean & Standard Deviation')
    plt.show(block=False)
```

This test support that the data is not stationary because the Test is greater than the p-value


*   Case 1: Both tests conclude that the series is not stationary -> series is not stationary
*   Case 2: Both tests conclude that the series is stationary -> series is stationary
*   Case 3: KPSS = stationary and ADF = not stationary  -> trend stationary, remove the trend to make series strict stationary
*   Case 4: KPSS = not stationary and ADF = stationary -> difference stationary, use differencing to make series stationary 

We have Case 1

```python
test_stationarity(ftd_gld['US dollars per troy'])
```

We must convert our data into stationary, so we are going to do  it by using:

SARIMAX enforce_stationarity=True


```python
p=d=q=range(0,2)

pdq = list(itertools.product(p,d,q))

seasonal_pdq = [(x[0],x[1],x[2],12) for x in list( itertools.product(p,d,q))]

mod= sm.tsa.statespace.SARIMAX(
          ftd_gld,
          order=(1, 1, 1),
          seasonal_order=(1, 1, 0, 12),
          enforce_stationarity = False
      )
resultados = mod.fit()

resultados.plot_diagnostics(figsize=(16,8))
plt.show()
```

```python
print('\n                 Upper and lower bounds by year.\n\n')
predicciones = resultados.get_prediction(start=pd.to_datetime('2004-12-31'))
pred_conf_int = predicciones.conf_int()
pred_conf_int.tail(10)
```

```python
prediccion_futurista = resultados.get_forecast(steps=5)
pred_ci=prediccion_futurista.conf_int()
ax = ftd_gld.plot(label='Pasado')
prediccion_futurista.predicted_mean.plot(ax=ax, label ='Pronostico', color = 'green', alpha =.8, figsize=(14,7))
ax.fill_between(pred_ci.index,
                pred_ci.iloc[:,0],
                pred_ci.iloc[:,1],
                color='blue',
                alpha=.1)
plt.legend()
plt.show()
```

#Holt Winters

```python
decompose_result = seasonal_decompose(ftd_gld['US dollars per troy'], model= 'multiplcative')
decompose_result.plot();
```

```python
ftd_gld.head(5)
```

```python
m = 12
alpha = 1/(2*m)
ftd_gld['HWES1'] = SimpleExpSmoothing(ftd_gld['US dollars per troy']).fit(smoothing_level=alpha,optimized=False,use_brute=True).fittedvalues
ftd_gld[['US dollars per troy','HWES1']].plot(title='Holt Winters Single Exponential Smoothing');
```

```python
ftd_gld['HWES2_ADD'] = ExponentialSmoothing(ftd_gld['US dollars per troy'],trend='add').fit().fittedvalues
ftd_gld['HWES2_MUL'] = ExponentialSmoothing(ftd_gld['US dollars per troy'],trend='mul').fit().fittedvalues
ftd_gld[['US dollars per troy','HWES2_ADD','HWES2_MUL']].plot(title='Holt Winters Double Exponential Smoothing: Additive and Multiplicative Trend');
```

```python
ftd_gld['HWES3_ADD'] = ExponentialSmoothing(ftd_gld['US dollars per troy'],trend='add',seasonal='add',seasonal_periods=12).fit().fittedvalues
ftd_gld['HWES3_MUL'] = ExponentialSmoothing(ftd_gld['US dollars per troy'],trend='mul',seasonal='mul',seasonal_periods=12).fit().fittedvalues
ftd_gld[['US dollars per troy','HWES3_ADD','HWES3_MUL']].plot(title='Holt Winters Triple Exponential Smoothing: Additive and Multiplicative Seasonality');
```

```python
train = ftd_gld[:100] 
test = ftd_gld[100:]
```

```python
fitted_model = ExponentialSmoothing(ftd_gld['US dollars per troy'],trend='mul',seasonal='mul',seasonal_periods=12).fit()
test_predictions = fitted_model.forecast(24)
train['US dollars per troy'].plot(legend=True,label='TRAIN')
test['US dollars per troy'].plot(legend=True,label='TEST',figsize=(14,7))
test_predictions.plot(legend=True,label='PREDICTION')
plt.title('Train, Test and Predicted Test using Holt Winters')
```
