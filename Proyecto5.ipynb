##### <font color='green'>Siempre la 1era parte es limpiar un poco nuestro dataset, y de esta forma nos quedaremos solo con la información que creemos será la más relevante para nuestro modelo.</font>

```python
# Analicemos los datos de la variable que queremos precedir:

sns.histplot(df['SalePrice']);
```

```python
# Para un análisis más exacto, necesitamos normaliar nuestra variable SalePrice:

df['SalePrice_Norm'] = np.log(df['SalePrice'])

sns.histplot(df['SalePrice_Norm']);
```

```python
# Eliminamos 'Id' ya que no nos aporta nada:

df.drop(columns='Id', inplace=True)
```

```python
# Eliminamos columnas con más del 80% de valores nulos:

porcentaje_nulos = df.isnull().sum() * 100 / df.shape[0]
valores_faltantes = pd.DataFrame({'column_name': df.columns, 'porcentaje_nulos': porcentaje_nulos})
valores_faltantes = valores_faltantes[valores_faltantes.porcentaje_nulos > 80].round(2)
columnas_eliminar = list(valores_faltantes.column_name)
df.drop(columns = columnas_eliminar, inplace = True)
df.head(3)
```

```python
# 1ero separaremos nuestras variables categóricas de las numéricas:

df_categoricas = df.select_dtypes(include=['object','category'])
df_numericas = df.select_dtypes(include='number')
```

```python
# 1ero trabajaremos nuestras variables categóricas.

df_categoricas.columns
```

```python
anova_categoricas = ols('SalePrice_Norm ~ MSZoning + Street +LotShape + LandContour + Utilities + LotConfig + LandSlope + Neighborhood+Condition1+Condition2+BldgType+HouseStyle+RoofStyle+RoofMatl+Exterior1st+Exterior2nd+MasVnrType+ExterQual+ExterCond+Foundation+BsmtQual+BsmtCond+BsmtExposure+BsmtFinType1+BsmtFinType2+Heating+HeatingQC+CentralAir+Electrical+KitchenQual+Functional+FireplaceQu+GarageType+GarageFinish+GarageQual+GarageCond+PavedDrive+SaleType+SaleCondition', data = df).fit()
sm.stats.anova_lm(anova_categoricas).head()
```

```python
confianza_categoricas = round((1 - sm.stats.anova_lm(anova_categoricas)[['PR(>F)']].sort_values(by = 'PR(>F)')).rename(columns = {'PR(>F)':'Confianza'})*100, 2)
confianza_categoricas[confianza_categoricas['Confianza'] == 100]

# Para ir disminuyendo nuestras columnas, nos quedaremos solo con las que tienen una confianza del 100%.
```

```python
# Arbitrariamente eliminaré Zona, ya que Neighborhood de por sí incluye la Zona.

columnas_relevantes_categoricas = ['Neighborhood','LotShape','BldgType','Exterior1st','ExterQual','RoofStyle','LandContour','MasVnrType','MasVnrType','BsmtExposure','RoofMatl',
                                   'BsmtQual','HouseStyle','KitchenQual','SaleCondition','Condition2','LotConfig','BsmtFinType1','Condition1','Exterior2nd']
print(columnas_relevantes_categoricas)
```

```python
for cat in columnas_relevantes_categoricas:
    print("\n",cat)
    print(df[cat].value_counts())
    print((df[cat].value_counts()/len(df[cat])))          
```

```python
categoricas_dummies = pd.get_dummies(df_categoricas, columns = ['Neighborhood'])
categoricas_dummies
```

```python
# Mi sentido común diría que la variable más importante DEBERÍA ser 'Neighborhood' (por experiencia buscando casas JAJAJA).

X = sm.add_constant(pd.get_dummies(df['Neighborhood']))
y = df['SalePrice_Norm']
model = sm.OLS(y, X).fit()
print_modelo1 = model.summary()
print(print_modelo1)
```

```python
# Para verificar lo anterior, nuevamente usaremos el sentido común, y veremos también por intuición si existe alguna de esas variables que tenga más peso en el valor del precio:

X = sm.add_constant(pd.get_dummies(df['SaleCondition']))
y = df['SalePrice_Norm']
model = sm.OLS(y, X).fit()
print_modelo1 = model.summary()
print(print_modelo1)
```

```python
X = sm.add_constant(pd.get_dummies(df['ExterQual']))
y = df['SalePrice_Norm']
model = sm.OLS(y, X).fit()
print_modelo1 = model.summary()
print(print_modelo1)
```
